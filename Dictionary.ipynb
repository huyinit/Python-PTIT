```python
# fromkeys()
#tạo một Dictionary từ một tập các keys với list đối tượng có thể thay đổi
keys = {'a', 'e', 'i', 'o', 'u' }
value = [1]
dict1 = dict.fromkeys(keys, value)
print(dict1)
value.append(2)
print(dict1)
```

```python
thisdict =	{  1: 4,  2: 5,  3: 6}
for x in thisdict:print(x,end=' ') 
print()
for x in thisdict.keys():print(x,end=' ')
print()
print(thisdict.keys())
```

```python
thisdict =	{  1: 4,  2: 5,  3: 6}

for x in thisdict: print(thisdict[x],end=' ')
print()
for x in thisdict.values():print(x,end=' ')
```

```python
thisdict =	{  1: 4,  2: 5,  3: 6}
# print key and value
for x, y in thisdict.items():print(x, y)
print(thisdict.items())
```

```python
d = {"b": 4, "c": 3, "a": 10}
a=dict(sorted(d.items(), key=lambda item:(item[0],item[1]) ))
print(a)
```

chuyển từ mảng 1 chiều về dicionary 

```python
a=[12, 43, 56, 14, 11, 11, 43, 43, 56, 14]
d={i:a.count(i) for i in a}
print(d)
```

chuyển từ dic vè mảng 2 chiều

```python
d={12: 1, 43: 3, 56: 2, 14: 2, 11: 2}
d=list(d.items())
d=[list(i) for i in d]
print(d)

print(*[i[1] for i in d])
val=list(set([i[1] for i in d]))
print(val)
```
