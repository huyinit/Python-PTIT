**2. enumerate**

```python
a=[1,2,3,4]
for i, val in enumerate(a):
    print('Index {} in array is: {}'.format(i, '%.5f'%val))
```

**3.Thêm**

```python
a=[1,2,3,4]
a.insert(0,0)# chèn bất kỳ (chỉ số, ký tự)
print(a)#[0, 1, 2, 3, 4]
a.append(5)#chèn vào cuối
print(a) #[0, 1, 2, 3, 4, 5]
```

**5.xóa**

```python
a=[1,2,3,4]
a.pop(3)#xóa bất kỳ
print(a) #[1, 2, 3]
a.remove(1)
print(a) #[2, 3]
```

**6.sao chép**

```python
a=[1,2,3,4]
b = a.copy()
c = a[:]
print(b)
print(c)
```

**7.sắp xếp**

```python
a=[1,2,3,4]
a.sort(reverse=True)#sx giảm dần
print(a)#[4, 3, 2, 1]
a.reverse()
print(a)
spam = ['a', 'z', 'A', 'Z']
spam.sort(key=str.lower)
print(spam)
a[0:3]=sorted(a[0:3],reverse=True)
print(a)
```

```python
a=[[1,2],[5,6],[3,7]]
a.sort()
print(a)
a=sorted(a,key=lambda l:l[1],reverse=True)
print(a)
print([i[0] for i in a])
print([i[1] for i in a])
```

**8.index và count**

```python
a=[1,2,3,4,1]
print(a.index(1)) # trả về vị trí của giá trị
print(a.count(1)) 
```

#10 chuyển từ chuỗi và từ điển sang list


```python
#chuyển từ chuỗi  sang list
s="123456789"
a1=list(s)
a2=[int(i) for i in s]
print(a1,a2,sep='\n')

#chuyển từ từ điển sang list
d={12: 1, 43: 3, 56: 2, 14: 2, 11: 2}
d=list(d.items())
print(d)
d=[list(i) for i in d]
print(d)
```
