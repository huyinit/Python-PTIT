```python
import datetime
def result1():
  strdate1="12:35" 
  strdate2="12:36" 
  time_in= datetime.datetime.strptime(strdate1, '%H:%M')
  time_out=datetime.datetime.strptime(strdate2, '%H:%M')
  dis= time_out-time_in
  return dis
def result2():
  strdate1="12:35" 
  strdate2="12:36" 
  time_in= datetime.datetime.strptime(strdate1, '%H:%M')
  time_out=datetime.datetime.strptime(strdate2, '%H:%M')
  dis= time_out-time_in
  return str(dis)
print(result1())
print(result2())
print(type(result1()))

```

```python
import datetime
def distanceOfTime():
  s1="1/1/2021:6/0/0" 
  s2="2/5/2022:7/0/0" 
  s1_datetime= datetime.datetime.strptime(s1, '%d/%m/%Y:%H/%M/%S')
  s2_datetime=datetime.datetime.strptime(s2, '%d/%m/%Y:%H/%M/%S')
  dis= s2_datetime-s1_datetime
  return str(dis)
print(distanceOfTime())
```
