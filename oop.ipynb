```python
#1 . Class không tham số + 3 phương thức __init__ ,__getitem__ ,__str__
class PerSon():
    def __init__(self):
        self.info = {
            'name':'huy init',
            'country':'Viet Nam',
            'number':123456789
        }
    def __getitem__(self,i):
        return self.info[i]
    def __str__(self):
        return "{} - {} - {}".format(self.info['name'],self.info['country'],self.info['number'])
p1 = PerSon()
print( p1['name'] )# Output: 'huy init'
print( p1['number'] )# Output: 123456789
print(p1)
```

```python
#2 . Class có tham số
# vd: rút gọn phân số
import math
class PS:
    def __init__(self, x,y):
        self.x = x
        self.y = y
    def pstg(self):
        temp=math.gcd(self.x,self.y)
        self.x =self.x/temp
        self.y =self.y/temp
    def __str__(self):
        return "{}/{}".format(int(self.x),int(self.y))  
ps=PS(40,100)
print(ps)
ps.pstg()
print(ps)
```

```python
# 3. truyền object vào trong 1 class
#vd: tính khoảng cách giữa 2 điểm tọa độ
import math
class Point:
    def __init__(self, x1,y1):
        self.x1 = x1
        self.y1 = y1
    def distance(self, Point):
        dist = math.sqrt((self.x1 - Point.x1)**2 
                         + (self.y1 - Point.y1)**2)
        return "%.4f"%dist
p1 = Point(1,1)
p2 = Point(0,5)
print(p1.distance(p2))
```

```python
# 4. viết kết quả trả về một đối tượng
#tính tổng phân số
import math
def object(tuso,mauso):
    return PS(tuso,mauso)
class PS:
    def __init__(self, x,y):
        self.x = x
        self.y = y    
    def pstg(self):
        temp=math.gcd(self.x,self.y)
        self.x =self.x/temp
        self.y= self.y/temp
    def add(self,PS ):
        mauso= self.y* PS.y 
        tuso= self.x*PS.y +self.y* PS.x
        return object(tuso,mauso)
    def __str__(self):
        return "{}/{}".format(int(self.x),int(self.y))  
p=PS(2,4)
q=PS(1,3)
res = p.add(q)
print(res)
res.pstg()
print(res)
```

```python
# 5. viết hàm tính toán 2 đối tượng
# vd tính tổng và tích của 2 số phức
class SoPhuc:
    def __init__(self,phanthuc,phanao):
        self.phanao = phanao
        self.phanthuc =phanthuc
    def __str__(self):
        a=abs(self.phanthuc)
        b=abs(self.phanao)
        if(self.phanao>0): return "{} + {}i".format(self.phanthuc, self.phanao)
        else:return "{} - {}i".format(self.phanthuc, b)

def add(sp1:SoPhuc,sp2:SoPhuc):
    a=sp1.phanthuc+sp2.phanthuc
    b= sp1.phanao+sp2.phanao
    return SoPhuc(a,b)
def multiplication(sp1:SoPhuc,sp2:SoPhuc):
    a=sp1.phanthuc*sp2.phanthuc-sp1.phanao*sp2.phanao
    b= sp1.phanao*sp2.phanthuc+sp1.phanthuc*sp2.phanao
    return SoPhuc(a,b)

sp1=SoPhuc(1,2) # 1+2i
sp2=SoPhuc(3,4) # 3+4i
sp3=add(sp1,sp2)
sp4=multiplication(sp1,sp2)
print(sp3,", ",sp4,sep="")
```

```python
# 6. tính toán về thời gian
import datetime
class person_play:
    def __init__(self, code, full_name, time_in, time_out):
        self.code=code
        self.full_name=full_name.strip()
        self.time_in= time_in
        self.time_out=time_out
        self.time=0
    def time_play(self):
        time_in= datetime.datetime.strptime(self.time_in, '%H:%M')
        time_out=datetime.datetime.strptime(self.time_out, '%H:%M')
        self.time= time_out-time_in
    def result(self):
        self.time=str(self.time)
        print(self.code, self.full_name, end=' ')
        a=[int(i)for i in self.time.split(':')]
        print("{} gio {} phut".format(a[0],a[1]))
    #note neu khong de str(self.time) thi kq se la datetime.delta
res=[]
per1=person_play("01", 'Nguyen Van An', '09:00', '10:30');res.append(per1)
per2=person_play("02", 'Hoang Van Nam', '09:00', '11:30');res.append(per2)
per3=person_play("03", 'Tran Hoa Binh', '09:00', '12:30');res.append(per3)
for i in range(3):res[i].time_play()
res.sort(key=lambda x:x.time,reverse=True)
for i in res: i.result()
```

```python
# 7.Kế thừa (Inheritance) 
class DaGiac:
     def __init__(self, socanh):
        self.n = socanh
        self.canh = [0 for i in range(socanh)]
     def nhapcanh(self):
        self.canh = [float(input("Bạn hãy nhập giá trị cạnh "+str(i+1)+" : ")) for i in range(self.n)]
     def hienthicanh(self):
        for i in range(self.n):
          print("Giá trị cạnh",i+1,"là",self.canh[i])
class TamGiac(DaGiac):
     def __init__(self):
        DaGiac.__init__(self,3)
     def dientich(self):
        a, b, c = self.canh
        s = (a + b + c) / 2
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('Diện tích của hình tam giác là %0.2f' %area)
t = TamGiac()
t.nhapcanh()
t.hienthicanh()
t.dientich()
```

```python
# 8. Overriding (Ghi đè) + (issubclass và isinstance)
class Base(): #nền tảng
	def __init__(self, x):
		self.x = x	
class Derived(Base): #nhận được từ, lấy được từ, tìm thấy nguồn gốc từ.
	def __init__(self, x, y):
		super(Derived, self).__init__(x)
		self.y = y
	def printXY(self):
	    print("x and y is:",self.x, self.y)
d = Derived(10, 200)
d.printXY()
print( issubclass(Base,Derived)) #class A là lớp con của class B?
print( isinstance(d,Derived))#obj là một instance của lớp Class hoặc là một instance của lớp con của Class.
```
